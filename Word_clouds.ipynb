```python
# import library
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from wordcloud import WordCloud
from os import listdir
import string
```

```python
file_path = "sport/"
```

```python
# read all the files and clean them for word-cloud
def process_files(file_path):
    all_tokens = ""
    for files in listdir(file_path):
        path = file_path+files
        file = open(path,"r")
        text = file.read()
        tokens = word_tokenize(text)
        print("words before the cleaning is {} in {}".format(len(tokens),files))
        tokens = [w for w in tokens if w not in string.punctuation]
        stop_words = set(stopwords.words("english"))
        tokoens = [w for w in tokens if w not in stop_words]
        tokens = [w for w in tokens if len(w)>3]
        print("words after the cleaning is {} in {}".format(len(tokens),files))
        text = " ".join(str(w) for w in tokens)
        all_tokens += text
    return all_tokens
    
    
```

```python
sports_data = process_files(file_path)
```

```python
sports_data[:200]
```

```python
# creating world cloud object
word_cloud = WordCloud(background_color="white",max_words=200).generate(sports_data)
```

```python
#ploting word-cloud
plt.figure(figsize=(20,20))
plt.imshow(word_cloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from PIL import Image
```

```python
football = np.array(Image.open("foot_ball.png"))
```

```python
football
```

```python
def transform(val):
    if val==0:
        return 255
    else:
        return val
```

```python
for i in range(len(football)):
    football[i]=list(map(transform,football[i]))
```

```python
football
```

```python
word_cloud = WordCloud(background_color="white",mask=football,contour_color="black",contour_width=0.3).generate(sports_data)
#ploting word-cloud
plt.figure(figsize=(20,20))
plt.imshow(word_cloud,interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
word_cloud.to_file("image.png")
```
